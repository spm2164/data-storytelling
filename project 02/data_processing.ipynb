```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
json = gpd.read_file("USA.json")
```

```python
weed_df = pd.read_csv("marijuana-street-price-clean.csv", index_col="date")
```

```python
json
```

```python
weed_df.index = pd.to_datetime(weed_df.index)
```

```python
weed_df.index=weed_df.index.sort_values()
```

```python
weed_df.sort_values(by="State")
```

```python
weed_df[weed_df["State"].isin(["Alabama"])].plot(y='HighQ')
```

```python
for state in weed_df["State"].unique():
    weed_df[weed_df["State"].isin([state])].plot(y='HighQ')
```

```python
for state_name in weed_df["State"].unique():
    state_df=weed_df[weed_df["State"].isin([state_name])]
    weed_dict=(state_df[['HighQ', 'MedQ', 'LowQ']].to_dict())
    weed_dict['state']=state_name
    for json_state in json["NAME"]:
        if (json_state==weed_dict['state']):
            print(json_state, weed_dict['state'])
```
